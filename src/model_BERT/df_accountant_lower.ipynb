```python
import pandas as pd
```

```python
# Load Bios dataset

df = pd.read_csv('accountant_complete.csv')
# drop the first column
df = df.drop(df.columns[1], axis=1)
df
```

```python
# Convert labels to 0 and 1
df['gender'] = df['gender'].apply(lambda x: 0 if x == 'Female' else 1)
df.head()
df.columns = ['text', 'labels']
df.head()
```

```python
# Lowercase the text
df["text"] = df["text"].str.lower()
df.head()
```

```python
df.to_csv("df_accountant_lower.csv", index=False)
```
