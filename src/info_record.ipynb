1. To retrieve gender distribution in full BiasBios Dataset.


```python
import pandas as pd

# Load the dataset (replace 'biasbios.csv' with your actual file)
df = pd.read_csv("BiasBias_DS.csv")

# Check column names
print(df.head())

```

```python
df.shape
```

```python
# As gender is stored as "gender" and contains "male" / "female"
df["gender"] = df["gender"].map({0: "male", 1: "female"})
gender_counts = df["gender"].value_counts()

# Print gender distribution
print("Gender Distribution:")
print(gender_counts)

# Convert to percentages
gender_percentages = gender_counts / gender_counts.sum() * 100
print("\nGender Distribution (Percentage):")
print(gender_percentages)
```

Visualization

```python
import matplotlib.pyplot as plt

gender_counts.plot(kind="bar", color=["blue", "pink"])
plt.title("Gender Distribution in BiasBios")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.show()
```

2. Analysing gender bias in profession..Gender Distribution in Professions.

```python
occupation_mapping = {
    0: "accountant",
    1: "architect",
    2: "attorney",
    3: "chiropractor",
    4: "comedian",
    5: "composer",
    6: "dentist",
    7: "dietitian",
    8: "dj",
    9: "filmmaker",
    10: "interior_designer",
    11: "journalist",
    12: "model",
    13: "nurse",
    14: "painter",
    15: "paralegal",
    16: "pastor",
    17: "personal_trainer",
    18: "photographer",
    19: "physician",
    20: "poet",
    21: "professor",
    22: "psychologist",
    23: "rapper",
    24: "software_engineer",
    25: "surgeon",
    26: "teacher",
    27: "yoga_teacher"
}
```

```python
df['profession'] = df['profession'].map(occupation_mapping)

```

```python
profession_gender_counts = df.groupby("profession")["gender"].value_counts(normalize=True).unstack()
print(profession_gender_counts.round(2))
```

```python
pip install seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(12, 6))

# Create heatmap
sns.heatmap(profession_gender_counts.T, cmap="coolwarm", annot=True, fmt=".2f")

# Labels and title
plt.title("Gender Representation in Professions")
plt.xlabel("Profession")
plt.ylabel("Gender")

# Show plot
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better visibility
plt.show()
```

** Number of biographies per Profession in whole dataset

```python
profession_counts = df.groupby("profession")['hard_text'].count()
print(profession_counts)
```

```python
# Bar Plot
# Sort values for better visualization
profession_counts_sorted = profession_counts.sort_values(ascending=False)

# Plot
plt.figure(figsize=(12, 6))
profession_counts_sorted.plot(kind="bar", color="skyblue")

# Labels and title
plt.title("Number of Biographies per Profession")
plt.xlabel("Profession")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")

# Show plot
plt.show()
```

```python
#Pie Chart
profession_counts_sorted.plot(kind="pie", autopct="%1.1f%%", figsize=(8, 8), colormap="Set2")

plt.title("Biographies Distribution by Profession")
plt.ylabel("")  # Hide y-axis label
plt.show()
```

6.  Bias-Bios dataset was collected from hundreds of thousands of online biographies, written in English, from the Common Crawl corpus. The Common Crawl Corpus is a vast dataset that contains petabytes of web data collected over 12 years, comprising over 50 billion web pages. It includes raw web page data, metadata extracts, and text extracts, and is available for public use.
